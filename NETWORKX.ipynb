Step6: Use from_pandas_edgelist() method from networkx package and create airline data graph

```python
data_1=data[['Origin','Dest']]
G=nx.Graph()

G=nx.from_pandas_edgelist(data_1,'Origin','Dest')

from matplotlib.pyplot import figure
figure(figsize=(10,5))
nx.draw_shell(G,with_labels=True)
```

### PART-4

Step13: Find shortest route between airports DAL and AUS based on AirTime by using dijkstra_path() method and print it

```python
G=nx.from_pandas_edgelist(data_1,'Origin','Dest')
print(nx.dijkstra_path(G,'DAL','AUS',weight='AirTime'))
#DAL - AUS is the shortest path for this source and destination
```

Step15: Find the length of shortest route between airports DAL and AUS based on AirTime by using dijkstra_path_length() method and print it

```python
#
for path in nx.all_shortest_paths(G,'DAL','AUS',weight='AirTime'):
  print(path)
#length of the shortest route
len(path)
```
